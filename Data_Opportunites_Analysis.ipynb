```python
def clean_job_title(job_title):

    if 'III' in job_title:
        job_title = job_title.replace('III', '')
        job_title = 'lead ' + job_title
    elif 'II' in job_title:
        job_title = job_title.replace('II', '')
        job_title = 'senior ' + job_title
    elif 'I' in job_title:
        job_title = job_title.replace('I', '')
        job_title = 'junior ' + job_title

    job_title = job_title.lower().replace('.', '').replace('sr', 'senior').replace('jr', 'junior').replace('/', ' ')

    if job_title.endswith('junior'):
        job_title = job_title.replace('junior', '')
        job_title = 'junior ' + job_title




    return job_title
```

```python
df['Job Title'] = df['Job Title'].apply(clean_job_title)
```

```python
def get_min_salary(minSalary):
  minSalary = minSalary.lower()
  minSalary = minSalary.split('-')[0]
  minSalary_num = int(''.join(list(filter(str.isdigit,minSalary))))

  if 'k' in minSalary:
    minSalary_num = minSalary_num *1000

  return minSalary_num
```

```python
def reven_handel(value):

  value = value.lower()

  if 'unknown' in value:
    return np.nan

  elif 'to' in value:
    # calculate average

    lower_val,upper_val = value.split('to')

    min_val = int(re.findall(r'\d+', lower_val)[0])
    max_val = int(re.findall(r'\d+', upper_val)[0])

    if 'billion' in lower_val:
      min_val = min_val*1000
    if 'billion' in upper_val:
      max_val = max_val*1000

    average_rev = int((min_val + max_val) / 2)

    return average_rev

  elif ('less' in value) or ('+' in value):
    number = int(re.findall(r'\d+', value)[0])

    if 'billion' in value:
      number*=10**3
    return number
```

```python
df['Revenue_cleaned']=df['Revenue'].apply(reven_handel)
```
